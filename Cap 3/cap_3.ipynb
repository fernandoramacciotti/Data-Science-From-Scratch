## matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams['figure.figsize'] = (12, 8)
```

### Grafico de linha

```python
years = [i for i in range(1950, 2020, 10)]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

plt.plot(years, gdp, color='green', marker='o', linestyle='solid')

# titulo
plt.title("GDP Nominal")

# label no eixo y
plt.ylabel("$ Bn")

# show
plt.show()
```

### Grafico de barras

```python
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]

xs = [i for i, _ in enumerate(movies)]

# barras com coordenadas x a esquerda [xs] e alturas num_oscars
plt.bar(xs, num_oscars)

plt.ylabel("# premios")
plt.title("Filmes")

# nomeia o eixo x com nomes dos filmes na barra central
plt.xticks([i for i, _ in enumerate(movies)], movies)

plt.show()
```

Grafico de barras como histograma

```python
from collections import Counter

grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]
decile = lambda grade: grade // 10 * 10
histogram = Counter(decile(grade) for grade in grades)

plt.bar([x for x in histogram.keys()],       
        histogram.values(),                      # altura de cada barra
        8)                                       # largura de 8 para cada barra

plt.axis([-5, 105, 0, 5])                        # eixo x de -5 a 105, eixo y de 0 a 5

plt.xticks([10 * i for i in range(11)])
plt.xlabel("Decil")
plt.ylabel("# alunos")
plt.title("Distribuicao das notas")
plt.show()
```

### Graficos de linhas (2)

```python
pwr_two = lambda x: 2 ** x
variance = [pwr_two(x) for x in range(9)]
bias_squared = sorted(variance, reverse=True)
total_error = [x + y for x, y in zip(variance, bias_squared)]
xs = [i for i, _ in enumerate(variance)]

# multiplas chamadas de plt.plot para mostrar multiplas series no mesmo grafico
plt.plot(xs, variance,      'g-',  label='variance')      # linha verde solida
plt.plot(xs, bias_squared,  'r-.', label='bias^2')        # linha vermelha com ponto tracejado
plt.plot(xs, total_error,   'b:',  label='total error')   # linha azul pontilhada

# como criamos label para as series, podemos adicionar legenda
# loc=9 significa top center
plt.legend(loc=9)
plt.xlabel("complexidade do modelo")
plt.title("Bias vs. Variance trade-off")
plt.show()
```
